```python
import pickle
import os
```

```python
pickle_dir = 'logging/pickle_data/'
file_names = os.listdir(pickle_dir)

def get_param_filename(n_hidden, lambdaval):
    return pickle_dir + 'n_{}-l_{}_params.pickle'.format(n_hidden, lambdaval)

def get_log_filename(n_hidden, lambdaval):
    return pickle_dir + 'n_{}-l_{}_log.pickle'.format(n_hidden, lambdaval)
```

```python
with open(get_param_filename(4, 0), 'rb') as f:
    params = pickle.load(f)
    
print(params)
```

```python
with open(get_log_filename(50, 0), 'rb') as f:
    log = pickle.load(f)
    
print(log)
```
